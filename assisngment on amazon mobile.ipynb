```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.amazon.in/s?k=oppo+phone&page=2&crid=1OTVS90MF6AVK&qid=1660107552&sprefix=%2Caps%2C611&ref=sr_pg_1'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(URL, headers=headers)

soup1 = BeautifulSoup(page.content, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")


```

```python
soup1
```

```python
soup2
```

```python
mobile = soup2.find_all('span',{'class':'a-size-medium a-color-base a-text-normal'})
```

```python
print(mobile)
```

```python
price = soup2.find_all('span',{'class':'a-price-whole'})
```

```python
print(price)
```

```python
ratings = soup2.find_all('i',{'class':'a-icon a-icon-star-small a-star-small-4 aok-align-bottom'})
```

```python
print(ratings)
```

```python
mobile_name = []
for i in range(0,len(mobile)):
    mobile_name.append(mobile[i].get_text())
```

```python
mob_name
```

```python
mob_name[:] = [mobile.lstrip('\n') for mobile in mob_name]
```

```python
mob_name
```

```python
mob_name[:] = [mobile.rstrip('\n') for mobile in mob_name]
```

```python
mob_name
```

```python
mob_price = []
for i in range(0,len(price)):
    mob_price.append(price[i].get_text())
```

```python
mob_price
```

```python
mob_price[:] = [price.lstrip('\n') for price in mob_price]
```

```python
mob_price
```

```python
mob_price[:] = [price.rstrip('\n') for price in mob_price]
```

```python
mob_price
```

```python
mob_ratings = []
for i in range(0,len(ratings)):
    mob_ratings.append(ratings[i].get_text())
```

```python
mob_ratings
```

```python
mob_ratings[:] = [ratings.lstrip('\n') for ratings in mob_ratings]
mob_ratings[:] = [ratings.rstrip('\n') for ratings in mob_ratings]
```
