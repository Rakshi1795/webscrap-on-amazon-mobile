```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.amazon.in/s?k=oneplus&crid=234SMV751ZZXP&sprefix=one%2Caps%2C316&ref=nb_sb_ss_pltr-ranker-retrain-acsession-opsacceptance_2_3'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(URL, headers=headers)

soup1 = BeautifulSoup(page.content, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
```

```python
soup1
```

```python
soup2
```

```python
mobile = soup2.find_all('span',{'class':'a-size-medium a-color-base a-text-normal'})
print(mobile)
```

```python
price = soup2.find_all('span',{'class':'a-price-whole'})
print(price)
```

```python
ratings = soup2.find_all('i',{'class':'a-icon a-icon-star-small a-star-small-4 aok-align-bottom'})
print(ratings)
```

```python
mobile_name = []
for i in range(0,len(mobile)):
    mobile_name.append(mobile[i].get_text())
    
mobile_name    
```

```python
mobile_name[:] = [mobile.lstrip('\n') for mobile in mobile_name]
mobile_name[:] = [mobile.rstrip('\n') for mobile in mobile_name]
mobile_name
```

```python
mob_price = []
for i in range(0,len(price)):
    mob_price.append(price[i].get_text())
    
mob_price    
```

```python
mob_price[:] = [price.lstrip('\n') for price in mob_price]
mob_price[:] = [price.rstrip('\n') for price in mob_price]
mob_price
```

```python
mobi_ratings = []
for i in range(0,len(ratings)):
    mobi_ratings.append(ratings[i].get_text())
    
mobi_ratings    
```

```python
mobi_ratings[:] = [ratings.lstrip('\n') for ratings in mobi_ratings]
mobi_ratings[:] = [ratings.rstrip('\n') for ratings in mobi_ratings]
mobi_ratings
```
